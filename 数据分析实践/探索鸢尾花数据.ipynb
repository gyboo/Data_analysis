```python
import pandas as pd
from numpy import nan
import numpy as np
#1.将数据集存成变量iris创建数据框的列名称['sepal_length',  'sepal_width', 'petal_length', 'petal_width', 'class'
iris = pd.read_csv('D:/datasets/mldata/iris.csv')
#iris.columns=['sepal_length','sepal_widt','petal_length','petal_widt','class'
iris.rename(columns={'species':'class'},inplace=True)
iris
```

```python
#查看缺失值
iris.isnull().sum()
```

```python
#将列petal_length的第10到19行设置为缺失值
iris.iloc[10:20,2] = nan
iris.isnull().sum()
```

```python
#将petal_lengt缺失值全部替换为1.0
iris = iris.fillna(1.0)
iris.isnull().sum()
```

```python
#删除列class
delete_iris = iris.drop('class',axis='columns')
delete_iris
```

```python
#将数据框前三行设置为缺失值
iris.iloc[0:3,0:4] = nan  #不同类型不能同时赋值
iris.iloc[0:3,4] = nan
iris.head()
```

```python
#删除有缺失值的行
drop_iris = iris.dropna(axis = 0)
drop_iris
```

```python
#重新设置索引
drop_iris = drop_iris.reset_index(drop=True)
drop_iris
```
