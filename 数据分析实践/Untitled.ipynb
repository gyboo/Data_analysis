```python
X = data['时间'].values.astype(float)
y = data['联合循环功率'].values.astype(float)
test_size = df['联合循环功率'].shape[0]//10
# 划分训练和测试集，最后12个值作为测试集
X_train = X[:-test_size]
X_test = X[-test_size:]
y_train = y[:-test_size]
y_test = y[-test_size:]

print(X_train.shape)
print(X_test.shape)
print(X_train[0:5])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))
X_train = scaler.fit_transform(X_train.reshape(-1, 1))
y_train = scaler.fit_transform(y_train.reshape(-1, 1))
X_train
```

```python
X_train = torch.FloatTensor(X_train).view(-1)
y_train = torch.FloatTensor(y_train).view(-1)
print(X_train.size())
```

```python
import torch
import torch.nn as nn

batch_size = 32
learning_rate = 1e-2
epoch = 5
class CNNnetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1d = nn.Conv1d(1,64,kernel_size=1)
        self.relu = nn.ReLU(inplace=True)
        self.fc1 = nn.Linear(64,50)
        self.fc2 = nn.Linear(50,1)
        
    def forward(self,x):
        # 该模型的网络结构为 一维卷积层 -> Relu层 -> Flatten -> 全连接层1 -> 全连接层2 
        x = self.conv1d(x)
        x = self.relu(x)
        x = x.view(-1)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
import time

torch.manual_seed(101)
model =CNNnetwork()

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
  

epochs = 10
model.train()
start_time = time.time()

for epoch in range(epochs):
    
    for i in range(0,7823):
        
        # 每次更新参数前都梯度归零和初始化
        optimizer.zero_grad()

        # 注意这里要对样本进行reshape，转换成conv1d的input size（batch size, channel, series length）
        y_pred = model(X_train[i].reshape(1,1,-1))
        loss = criterion(y_pred, y_train[i])
        loss.backward()
        optimizer.step()
        
    print(f'Epoch: {epoch+1:2} Loss: {loss.item():10.8f}')
    
print(f'\nDuration: {time.time() - start_time:.0f} seconds')
```

```python
future = 64
# 选取序列最后12个值开始预测
preds = X_train[-64:].tolist()
print(len(preds))
# 设置成eval模式
model.eval()

# 循环的每一步表示向时间序列向后滑动一格
for i in range(future):
    seq = torch.FloatTensor(preds[-64:])
    with torch.no_grad():
        preds.append(model(seq.reshape(1,1,-1)).item())

# 逆归一化还原真实值
true_predictions = scaler.inverse_transform(np.array(preds[64:]).reshape(-1, 1))
print(true_predictions.shape)

print(len(preds))
```
